```python
import os
import sys
from pycocotools.coco import COCO

#instances
instances_annFile = "cocoapi/annotations/instances_val2017.json"
coco = COCO(instances_annFile)

#captions
captions_annFile = "cocoapi/annotations/captions_val2017.json"
coco_caps = COCO(captions_annFile)

# get image ids 
idx = list(coco.anns.keys())
```

```python
import numpy as np
import matplotlib.pyplot as plt
import skimage.io as io
%matplotlib inline
```

```python
rand = np.random.choice(idx)
image_number = coco.anns[rand]["image_id"]
image = coco_caps.loadImgs(image_number)[0]
url = image['coco_url']
img = io.imread(url)
plt.imshow(img)
plt.show()
annIds = coco_caps.getAnnIds(imgIds=image_number);
anns = coco_caps.loadAnns(annIds)
coco_caps.showAnns(anns)
```
